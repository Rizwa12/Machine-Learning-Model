**Build Model Using Relu Function Tanh regularization and dropout**

**1.Create Model Using Relu Function**

```
# we are passing activation function as a parameter here so that we can call this function with tanh or relu while
# fitting and training the model

from keras import models
from keras import layers
def build_model():
  model = models.Sequential()
  model.add(layers.Dense(10, activation= 'relu',input_shape=(train.shape[1],)))
  model.add(layers.Dense(8, activation= 'relu'))
  model.add(layers.Dense(6, activation= 'relu'))
  model.add(layers.Dense(1))
  model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
  return model
```

```
build_model().summary()
```

**2. Create Model Using Tanh Function**

```
from keras import models
from keras import layers
def build_model_tanh():
  model = models.Sequential()
  model.add(layers.Dense(10,activation='tanh',input_shape=(train.shape[1],)))
  model.add(layers.Dense(8,activation='tanh'))
  model.add(layers.Dense(6, activation='tanh'))
  model.add(layers.Dense(1))
  model.compile(optimizer='rmsprop',loss='mse',metrics=['mae'])
  return model
```

```
build_model_tanh().summary()
```

**3. Build Regularized Model**

```
from keras import regularizers
def build_model_regular():
  model = models.Sequential()
  model.add(layers.Dense(10,activation='tanh',kernel_regularizer=regularizers.l1_l2(l1=0.001,l2=0.001),input_shape=(train.shape[1],)))
  model.add(layers.Dense(8,activation='tanh',kernel_regularizer=regularizers.l1_l2(l1=0.001,l2=0.001)))
  model.add(layers.Dense(6, activation='tanh',kernel_regularizer=regularizers.l1_l2(l1=0.001,l2=0.001)))
  model.add(layers.Dense(1))
  model.compile(optimizer='rmsprop',loss='mse',metrics=['mae'])
  return model
```

```
build_model_regular().summary()
```

**4. build dropout model**

```
from keras import regularizers
def build_model_drop():
  model=models.Sequential()
  model.add(layers.Dense(10,activation='relu',input_shape=(train.shape[1],)))
  model.add(layers.Dropout(0.5))
  model.add(layers.Dense(8, activation= 'relu'))
  model.add(layers.Dropout(0.5))
  model.add(layers.Dense(6, activation= 'relu'))
  model.add(layers.Dropout(0.5))
  model.add(layers.Dense(1))
  model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
  return model
```

**K Fold Validation Step**

**1.K Fold Validation for Relu**

```
#149/4
import numpy as np
k=4
num_val_samples= len(train)//4
num_epochs=100
all_scores=[]
for i in range(k):
  print('processing fold #',i)
  val_data = train[i* num_val_samples:(i+1)*num_val_samples]
  val_targets = trian_l[i*num_val_samples:(i+1)*num_val_samples]
  partial_train_data = np.concatenate([train[:i*num_val_samples],train[(i+1)*num_val_samples:]],axis=0)
  partial_train_target = np.concatenate([trian_l[:i * num_val_samples],trian_l[(i+1)* num_val_samples:]],axis=0)
  model = build_model()
  model.fit(partial_train_data,partial_train_target,epochs=num_epochs,batch_size=1,verbose=1)
  val_mse, vale_mae = model.evaluate(val_data,val_targets,verbose=0)
  all_scores.append(vale_mae)
```
